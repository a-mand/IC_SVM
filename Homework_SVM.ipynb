**1. Cálculo do Custo Original (SVM)**

$$
f(\mathbf{x}) = \sum_{i=1}^{450} \alpha_i (\mathbf{s}_i \cdot \mathbf{x}) + b
$$

- Dimensão dos vetores K = 5
- Número de vetores de suporte N = 450
- Operações por produto interno:
  - Multiplicações: 5 (para cada feature)
  - Adições: 4 (soma de cada multiplicação)

Cada vetor de suporte requer: [N * 5 (multiplicações)] + [N * 4 (adições)] 

Para N = 450 vetores:
(450 * 5) + (450 * 4) = 2250 + 1800 = 4050 operações

Multiplicação do coeficiente $\alpha_i$ = 450 * 1 = 450 (1 multiplicação por vetor de suporte)

Somando os resultados de $\alpha_i (\mathbf{s}_i \cdot \mathbf{x})$ = 499
Adicionando o bias = 1

    499 + 1 = 450

Multiplicações: 2250 (produto interno) + 450 (multiplicação por $\alpha_i$) = 2700

Adições: 1800 (produto interno) + 450 (soma dos termos e bias) = 2250

Custo Original (SVM) = 2700 + 2250 = 4950 operações

**2. Cálculo do Custo como Perceptron**
   
A conversão para perceptron requer apenas: [5 (multipliações) + 4 (adições)] + 1 (bias) = 10 operações

**3. Fator de Redução**

F = Coriginal/Cperceptron = 4950/10 = 495

**Resposta Final:**

    * O fator de redução de custo computacional é mais de 495 vezes.
    * Entende-se que a conversão para Perceptron reduz drasticamente o custo (em 495 vezes).

   ```python
   w = Σ (αᵢyᵢ) * xᵢ
     = -0.45994152 * [0., -4.] + -0.27992202 * [-1., 2.] + 0.73986354 * [-2., -2.]
     = [-1.19980506, -0.19980506]

**Função de decisão perceptron**
$$
f(z) = w \cdot z + b
$$

$$
f(z) = -1.19980506 \cdot z_1 - 0.19980506 \cdot z_2 - 1.79954513
$$

Custo da SVM Original:
Para cada predição:
- Calcula 3 produtos escalares (z com cada vetor de suporte)
- 3 multiplicações pelos coeficientes αy
- 2 adições (somatório)
- 1 adição do bias

**Total:** 6 multiplicações + 3 adições

Custo como Perceptron:
Para cada predição:
- 2 multiplicações (w · z)
- 1 adição (com bias)

**Total:** 2 multiplicações + 1 adição

Economia:
- Redução de 66% nas multiplicações (6 → 2)
- Redução de 66% nas adições (3 → 1)
- Eliminação da necessidade de armazenar vetores de suporte para predição

## Questão 6

````python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

def load_data(filename):
    """Carrega dados do arquivo e separa em features (X) e labels (y)"""
    data = np.loadtxt(filename, delimiter=',')
    X = data[:, :-1]  # Todas as colunas exceto a última
    y = data[:, -1]    # Última coluna
    return X, y


X_train, y_train = load_data('datasets/dataset_train.txt')
X_test, y_test = load_data('datasets/dataset_test.txt')

# Normalização dos dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test)      

# Treinar SVM com kernel linear (C=1)
svm = SVC(kernel='linear', C=1)
svm.fit(X_train_scaled, y_train)

# Avaliar desempenho no teste
accuracy = svm.score(X_test_scaled, y_test)
print(f"Acurácia no teste: {accuracy:.2%}")

# Converter SVM para Perceptron
w = svm.coef_[0]      # Pesos
b = svm.intercept_[0]  # Bias

def perceptron_predict(X):
    """Função de predição do perceptron equivalente"""
    return np.where(X.dot(w) + b > 0, 1, 0)

# Verificar se as predições são iguais
svm_predictions = svm.predict(X_test_scaled)
perceptron_predictions = perceptron_predict(X_test_scaled)

print("\nPrevisões SVM:", svm_predictions)
print("Previsões Perceptron:", perceptron_predictions)
print("São iguais?", np.all(svm_predictions == perceptron_predictions))

# Análise de custo computacional
print("\nAnálise de custo computacional:")
print(f"Dimensão dos dados: {X_train.shape[1]} features")
print(f"Memória SVM: Armazena {len(w)} pesos + 1 bias = {len(w)+1} valores")
print(f"Memória Perceptron: Armazena {len(w)} pesos + 1 bias = {len(w)+1} valores")
print("Operações por predição: produto interno (d multiplicações + d somas) + comparação")
````

Não há tanta diferença entre SVM linear e Perceptron:
* Ambas usam a mesma quantidade de memória 
* Realizam as mesmas operações ( produto interno w ⋅ x = b)

A única diferença é que o perceptron é mais simples de implementar.

#### letra a)

````python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

def load_data(filename):
    """Carrega dados do arquivo e separa em features (X) e labels (y)"""
    data = np.loadtxt(filename, delimiter=',')
    X = data[:, :-1]  
    y = data[:, -1]    
    return X, y

# Carregar dados
X_train, y_train = load_data('datasets/dataset_train.txt')
X_val, y_val = load_data('datasets/dataset_validation.txt')
X_test, y_test = load_data('datasets/dataset_test.txt')

# Normalização dos dados (baseado no treino)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Valores de C e gamma para testar
C_values = [0.01, 1, 100]
gamma_values = [0.5, 1]

# Armazenar resultados
best_accuracy = 0
best_params = {}
results = []  # Lista para armazenar todos os resultados

# Testar todas as combinações
for C in C_values:
    for gamma in gamma_values:
        # Treinar SVM com RBF
        svm = SVC(kernel='rbf', C=C, gamma=gamma)
        svm.fit(X_train_scaled, y_train)
        
        # Avaliar no conjunto de validação
        val_accuracy = svm.score(X_val_scaled, y_val)
        # Avaliar no conjunto de teste
        test_accuracy = svm.score(X_test_scaled, y_test)
        
        # Armazenar resultados
        results.append({
            'C': C,
            'gamma': gamma,
            'val_accuracy': val_accuracy,
            'test_accuracy': test_accuracy
        })
        
        print(f"C={C}, gamma={gamma} → Val: {val_accuracy:.2%} | Teste: {test_accuracy:.2%}")
        
        # Verificar se é o melhor modelo até agora (baseado na validação)
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            best_params = {'C': C, 'gamma': gamma}
            best_svm = svm

# Resultados de todas as combinações
print("\nResultados completos:")
print("C\tgamma\tValidação\tTeste")
for res in results:
    print(f"{res['C']}\t{res['gamma']}\t{res['val_accuracy']:.2%}\t\t{res['test_accuracy']:.2%}")

# Melhor modelo
print("\nMelhor combinação de hiperparâmetros (baseado na validação):")
print(f"C={best_params['C']}, gamma={best_params['gamma']}")
print(f"Acurácia no conjunto de validação: {best_accuracy:.2%}")
print(f"Acurácia no conjunto de teste: {best_svm.score(X_test_scaled, y_test):.2%}")
````

A SVM com a melhor perfonamance dno conjunto de validação é C=0.01 e gamma=0.5 

- gamma = 0,5: generaliza mais, evita overfitting

- C = 0,01: suficiente para alcançar a melhor acurácia

````python
# Selecionar os dois primeiros exemplos do conjunto de teste
z1 = X_test_scaled[0]
z2 = X_test_scaled[1]

# 1. Calcular usando decision_function() do scikit-learn
scores_sklearn = best_svm.decision_function([z1, z2])
print(f"Scores do scikit-learn (decision_function): {scores_sklearn}")

# 2. Calcular manualmente usando os parâmetros do SVM
def rbf_kernel(x1, x2, gamma):
    return np.exp(-gamma * np.linalg.norm(x1 - x2)**2)

def manual_decision_function(x, svm_model):
    # Obter parâmetros do modelo
    support_vectors = svm_model.support_vectors_
    dual_coef = svm_model.dual_coef_[0]  # Para classificação binária
    bias = svm_model.intercept_[0]
    gamma = svm_model.gamma
    
    # Calcular a soma dos termos do kernel
    kernel_sum = 0
    for i in range(len(support_vectors)):
        kernel_sum += dual_coef[i] * rbf_kernel(support_vectors[i], x, gamma)
    
    return kernel_sum + bias

# Calcular para z1 e z2
score_z1_manual = manual_decision_function(z1, best_svm)
score_z2_manual = manual_decision_function(z2, best_svm)

print(f"\nScores calculados manualmente:")
print(f"z1: {score_z1_manual}")
print(f"z2: {score_z2_manual}")

# Comparação numérica
print("\nDiferenças absolutas:")
print(f"z1: {abs(scores_sklearn[0] - score_z1_manual)}")
print(f"z2: {abs(scores_sklearn[1] - score_z2_manual)}")
````
